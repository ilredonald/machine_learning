#La régression linéaire avec scikit-learn

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.datasets import make_regression
from sklearn.linear_model import SGDRegressor
from sklearn.preprocessing import PolynomialFeatures
from sklearn.model_selection import train_test_split
```

```python
# Création du Dataset.
np.random.seed(0)

x, y = make_regression(n_samples = 500, n_features = 1, noise = 5)
y = y**3

poly_features = PolynomialFeatures(degree = 3, include_bias = False)
x = poly_features.fit_transform(x)

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
print(x.shape)
print(x_train.shape)
print(x_test.shape)

plt.scatter(x[:,0],y)
```

```python
# Création du modèle.
model = SGDRegressor(max_iter = 1000, eta0 = 1e-3)

#entrainement du modèle.
model.fit(x_train, y_train)
print(f"Coefficient de détermination R2 = {model.score(x_test, y_test)}")
```

```python
#prediction 
plt.scatter(x[:,0],y)
plt.scatter(x[:,0], model.predict(x), marker="*", s = 10, c="red")
```
